## Task 9: Common Elements
Given
list1 = [10,20,30,40]
list2 = [30,40,50,60]
Find common elements using:
Lists only
Sets

```python
# Create two lists
list1 = [10, 20, 30, 40]
list2 = [30, 40, 50, 60]

# Create an empty list to store common elements
common_elements = []

# Find common elements using lists
for number in list1:
    if number in list2:
        common_elements.append(number)

print("Using lists:", common_elements)

# Convert the lists into sets
set1 = set(list1)
set2 = set(list2)

# Find common elements using sets
print("Using sets:", set1 & set2)
```

## Task 10: Character Frequency
Take a string input from the user.
Example
banana
Output
{
'a':3,
'b':1,
'n':2
}

```python
# Read a string from the user
word = input("Enter a string: ")

# Create an empty dictionary
character_count = {}

# Count the frequency of each character
for character in word:
    if character not in character_count:
        character_count[character] = 1
    else:
        character_count[character] += 1

# Display the character frequencies
print(character_count)
```

## Task 15: Nested Dictionary
Create
students = {
   "John":{
       "Math":90,
       "Science":85
   },
   "Alice":{
       "Math":95,
       "Science":92
   }
}
Print:
John's Math marks
Alice's Science marks
Average marks of John

```python
# Create a nested dictionary
students = {
    "John": {
        "Math": 90,
        "Science": 85
    },
    "Alice": {
        "Math": 95,
        "Science": 92
    }
}

# Display John's Math marks
print("John's Math marks:", students["John"]["Math"])

# Display Alice's Science marks
print("Alice's Science marks:", students["Alice"]["Science"])

# Calculate John's average marks
average = sum(students["John"].values()) / len(students["John"])

print("Average marks of John:", average)
```

## Task 18: Separate Even and Odd Numbers
Given
nums = [12,5,8,17,20,3,14]
Create two separate lists:
Even numbers
Odd numbers
Then convert both into tuples.

```python
# Create a list of numbers
numbers = [12, 5, 8, 17, 20, 3, 14]

# Create a list of even numbers
even_numbers = [number for number in numbers if number % 2 == 0]

# Create a list of odd numbers
odd_numbers = [number for number in numbers if number % 2 == 1]

# Convert both lists into tuples
even_tuple = tuple(even_numbers)
odd_tuple = tuple(odd_numbers)

# Display the tuples
print("Tuple of even numbers:", even_tuple)
print("Tuple of odd numbers:", odd_tuple)
```

## Task 19: Inventory Management
Create a dictionary like
inventory = {
   "Pen":25,
   "Book":10,
   "Pencil":40
}
Menu:
1. Add item
2. Update quantity
3. Delete item
4. Search item
5. Print inventory
6. Exit
Use dictionary methods appropriately.

```python
# Create an inventory dictionary
inventory = {
    "Pen": 25,
    "Book": 10,
    "Pencil": 40
}

# Display the menu repeatedly
while True:
    print("\nMenu")
    print("1. Add item")
    print("2. Update quantity")
    print("3. Delete item")
    print("4. Search item")
    print("5. Print inventory")
    print("6. Exit")

    choice = int(input("Enter your choice: "))

    # Add a new item
    if choice == 1:
        item = input("Enter item name: ")
        quantity = int(input("Enter quantity: "))
        inventory[item] = quantity
        print("Item added successfully.")
        print(inventory)

    # Update item quantity
    elif choice == 2:
        item = input("Enter item name to update: ")

        if item in inventory:
            quantity = int(input("Enter new quantity: "))
            inventory[item] = quantity
            print("Quantity updated successfully.")
            print(inventory)
        else:
            print("Item not found.")

    # Delete an item
    elif choice == 3:
        item = input("Enter item name to delete: ")

        if item in inventory:
            del inventory[item]
            print("Item deleted successfully.")
            print(inventory)
        else:
            print("Item not found.")

    # Search an item
    elif choice == 4:
        item = input("Enter item name to search: ")

        if item in inventory:
            print("Item found.")
        else:
            print("Item not found.")

    # Display inventory
    elif choice == 5:
        print(inventory)

    # Exit the program
    elif choice == 6:
        print("Exiting...")
        break

    else:
        print("Invalid choice.")
```

## Task 20: Employee Database
Create a program that stores employee details.
Each employee should have:
{
   "ID":101,
   "Name":"Alice",
   "Department":"HR",
   "Salary":50000
}
Store multiple employees in a dictionary or list.
Implement:
Add employee
Delete employee
Update salary
Search by ID
Display highest-paid employee
Display average salary
Display all employees sorted by salary

```python
# Create an empty list to store employee records
employees = []


# Function to add a new employee
def add_employee():
    employee_id = int(input("Enter Employee ID: "))
    employee_name = input("Enter Employee Name: ")
    department = input("Enter Department: ")
    salary = float(input("Enter Salary: "))

    employee = {
        "ID": employee_id,
        "Name": employee_name,
        "Department": department,
        "Salary": salary
    }

    employees.append(employee)
    print("Employee added successfully!")


# Function to delete an employee using ID
def delete_employee():
    employee_id = int(input("Enter Employee ID to delete: "))

    for employee in employees:
        if employee["ID"] == employee_id:
            employees.remove(employee)
            print("Employee deleted successfully!")
            return

    print("Employee not found!")


# Function to update an employee's salary
def update_salary():
    employee_id = int(input("Enter Employee ID to update salary: "))

    for employee in employees:
        if employee["ID"] == employee_id:
            new_salary = float(input("Enter the new salary: "))
            employee["Salary"] = new_salary
            print("Salary updated successfully!")
            return

    print("Employee not found!")


# Function to search an employee by ID
def search_by_id():
    employee_id = int(input("Enter Employee ID to search: "))

    for employee in employees:
        if employee["ID"] == employee_id:
            print("Employee found:")
            print(employee)
            return

    print("Employee not found!")


# Function to display the highest-paid employee
def highest_paid():
    if employees:
        highest = max(employees, key=lambda employee: employee["Salary"])
        print("Highest Paid Employee:")
        print(highest)
    else:
        print("No employees in the database!")


# Function to calculate the average salary
def average_salary():
    if employees:
        average = sum(employee["Salary"] for employee in employees) / len(employees)
        print("Average Salary:", average)
    else:
        print("No employees in the database!")


# Function to display employees sorted by salary
def sort_by_salary():
    if employees:
        sorted_employees = sorted(employees, key=lambda employee: employee["Salary"])

        print("Employees sorted by salary:")

        for employee in sorted_employees:
            print(employee)
    else:
        print("No employees in the database!")


# Display the menu until the user chooses to exit
while True:
    print("\n===== Employee Database Menu =====")
    print("1. Add Employee")
    print("2. Delete Employee")
    print("3. Update Salary")
    print("4. Search by ID")
    print("5. Highest Paid Employee")
    print("6. Average Salary")
    print("7. Sort by Salary")
    print("8. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        add_employee()

    elif choice == 2:
        delete_employee()

    elif choice == 3:
        update_salary()

    elif choice == 4:
        search_by_id()

    elif choice == 5:
        highest_paid()

    elif choice == 6:
        average_salary()

    elif choice == 7:
        sort_by_salary()

    elif choice == 8:
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Please try again.")
```
