## Task 1: Even or Odd
Write a program that takes a number as input and checks whether it is even or odd.
Example:
Input: 8
Output: Even

```python
# Take a number as input from the user
num = int(input("Enter a number to check even or odd:"))

# Check whether the number is even or odd
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

## Task 2: Largest of Three Numbers
Take three numbers from the user and print the largest among them using conditional statements.
Example:
Input:
10
25
18

Output:
25 is the largest number.

```python
# Read three numbers from the user
a = int(input("Enter 1st number:"))
b = int(input("Enter 2nd number:"))
c = int(input("Enter 3rd number:"))

# Compare the numbers and print the largest
if a >= b and a >= c:
    print(a, "is the largest")
elif b >= a and b >= c:
    print(b, "is the largest")
else:
    print(c, "is the largest")
```

## Task 3: Multiplication Table
Take a number from the user and print its multiplication table from 1 to 10 using a for loop.
Example:
Input: 5

Output:
5 x 1 = 5
5 x 2 = 10
...
5 x 10 = 50

```python
# Read a number from the user
num = int(input("Enter a number:"))

# Print multiplication table from 1 to 10
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

## Task 4: Count Positive, Negative, and Zero
Ask the user to enter 10 numbers using a loop.
At the end, display:
Number of positive numbers
Number of negative numbers
Number of zeros
Example:
Input:
5
-2
0
8
-7
...

Output:
Positive: 4
Negative: 3
Zero: 3

```python
# Initialize counters
positive = 0
negative = 0
zero = 0

# Read 10 numbers
print("Enter 10 numbers one by one")
for i in range(1, 11):
    num = int(input(f"Enter number {i}:"))

    # Update the respective counter
    if num > 0:
        positive += 1
    elif num < 0:
        negative += 1
    else:
        zero += 1

# Display the results
print("Positive:", positive)
print("Negative:", negative)
print("Zero:", zero)
```

## Task 5: Password Checker (Using while)
Keep asking the user to enter the password until they enter:
python123
If the password is correct, print:
Access Granted
Otherwise print:
Wrong Password. Try Again.

```python
# Store the correct password
secret = "python123"

# Keep asking until the correct password is entered
while True:
    password = input("Enter the correct password: ")

    if password == secret:
        print("Access Granted")
        break

    print("Wrong Password! Try Again.")
```

## Task 6: Skip Multiples of 3 (Using continue)
Print numbers from 1 to 30, but skip all multiples of 3 using the continue statement.
Expected Output:
1
2
4
5
7
8
10
...
29

```python
# Print numbers from 1 to 30
for i in range(1, 31):

    # Skip multiples of 3
    if i % 3 == 0:
        continue

    print(i,end=" ")
```

## Task 7: Stop at First Negative Number (Using break)
Ask the user to enter numbers continuously.
If the user enters a negative number, stop taking input using break.
Finally print:
Total numbers entered (excluding the negative one)
Sum of all entered numbers
Example:
Input:
5
8
10
3
-2

Output:
Count = 4
Sum = 26

```python
# Initialize variables
total = 0
count = 0

# Read numbers until a negative number is entered
while True:
    num = int(input("Enter number:"))

    # Stop when a negative number is entered
    if num < 0:
        break

    total += num
    count += 1

# Display the total and count
print("Sum:", total)
print("Count:", count)
```

## Task 8: Find a Number in a List
Create a list of numbers.
Ask the user to enter a number to search.
Use a loop and break to stop once the number is found.
Print:
Number Found
Otherwise:
Number Not Found

```python
# Create a sample list
numbers = [1, 3, 46, 77]

# Read the number to search
num = int(input("Enter a number to search:"))

# Search for the number
for n in numbers:
    if n == num:
        print("Number Found")
        break
else:
    print("Number Not Found")
```

## Task 10: Number Guessing Game
Store a secret number (for example, 17).
Ask the user to guess the number repeatedly.
If the guess is smaller, print:
 Too Low

If greater:
 Too High

If correct:

 Congratulations! You guessed it. and stop the loop using break.

```python
# Store the secret number
secret = 56

# Keep asking until the correct number is guessed
while True:
    guess = int(input("Guess a number: "))

    if guess < secret:
        print("Too Low")
    elif guess > secret:
        print("Too High")
    else:
        print("Correct Guess!")
        break
```
