## Task 1: Safe Integer Input
### Problem:

- Ask the user to enter an integer.
- Convert the input to an integer using `int()`.
- Print the entered number.
- Handle invalid input using `ValueError`.

```python
# Read an Integer using Exception Handling

try:
    # Read an integer from the user
    number = int(input("Enter an integer: "))

    # Display the entered integer
    print("You entered", number)

# Handle invalid integer input
except ValueError:
    print("Invalid input! Please enter an integer.")
```

## Task 2: Safe Division
### Problem:

- Ask the user to enter a dividend and a divisor.
- Perform the division.
- Display the division result.
- Handle invalid input using `ValueError`.
- Handle division by zero using `ZeroDivisionError`.

```python
# Perform Division using Exception Handling

try:
    # Read the dividend and divisor from the user
    n1 = int(input("Enter dividend: "))
    n2 = int(input("Enter divisor: "))

    # Perform the division
    division = n1 / n2

    # Display the division result
    print("Division result:", division)

# Handle invalid numeric input
except ValueError:
    print("Invalid input! Please enter valid numbers.")

# Handle division by zero
except ZeroDivisionError:
    print("Cannot divide by zero.")
```

## Task 3: String to Integer Addition
### Problem:

- Ask the user to enter two numbers.
- Convert both inputs to integers.
- Add the two numbers.
- Display the sum.
- Handle invalid input using `ValueError`.

```python
# Add Two Numbers using Type Conversion

try:
    # Read two numbers as strings
    num1 = input("Enter 1st number: ")
    num2 = input("Enter 2nd number: ")

    # Convert the strings to integers and add them
    result = int(num1) + int(num2)

    # Display the sum
    print("Sum =", result)

# Handle invalid numeric input
except ValueError:
    print("Invalid input! Please enter only numbers.")
```

## Task 4: Temperature Converter
### Problem:

- Ask the user to enter the temperature in Celsius.
- Convert the temperature to Fahrenheit.
- Display the converted temperature.
- Handle invalid input using `ValueError`.

```python
# Temperature Converter using Exception Handling

try:
    # Read the temperature in Celsius
    celsius = float(input("Enter temperature in Celsius: "))

    # Convert Celsius to Fahrenheit
    fahrenheit = (celsius * 9/5) + 32

    # Display the converted temperature
    print(f"{fahrenheit} °F")

# Handle invalid numeric input
except ValueError:
    print("Invalid input! Please enter a valid number.")
```

## Task 5: Student Marks Validator
### Problem:

- Ask the user to enter marks between 0 and 100.
- Convert the input to an integer.
- Check whether the marks are within the valid range.
- Raise a `ValueError` if the marks are outside the valid range.
- Display a success message for valid marks.

```python
# Validate Student Marks using Exception Handling

try:
    # Read the student's marks
    marks = int(input("Enter marks (0–100): "))

    # Check whether the marks are within the valid range
    if 0 <= marks <= 100:
        print("Valid Marks")
    else:
        # Raise an exception if the marks are outside the valid range
        raise ValueError("Marks should be between 0 and 100.")

# Handle invalid input or invalid marks
except ValueError as e:
    print(e)
```


## Task 6: Mini Calculator with Exception Handling
### Problem:

- Ask the user to enter two numbers.
- Ask the user to choose an arithmetic operator (`+`, `-`, `*`, `/`).
- Perform the selected operation.
- Display the result.
- Handle invalid numeric input using `ValueError`.
- Handle division by zero using `ZeroDivisionError`.
- Handle invalid operators using `Exception`.

```python
# Mini Calculator with Exception Handling

try:
    # Read the first number from the user
    num1 = float(input("Enter first number: "))

    # Read the arithmetic operator
    operator = input("Enter operator (+, -, *, /): ")

    # Read the second number from the user
    num2 = float(input("Enter second number: "))

    # Perform the selected arithmetic operation
    if operator == "+":
        result = num1 + num2
    elif operator == "-":
        result = num1 - num2
    elif operator == "*":
        result = num1 * num2
    elif operator == "/":
        # Check for division by zero
        if num2 == 0:
            raise ZeroDivisionError("Division by zero is not allowed.")
        result = num1 / num2
    else:
        # Raise an exception for an invalid operator
        raise ValueError("Invalid operator entered.")

# Handle invalid number input or invalid operator
except ValueError as ve:
    print("Error:", ve)

# Handle division by zero
except ZeroDivisionError as zde:
    print("Error:", zde)

# Handle any other unexpected errors
except Exception as e:
    print("Unexpected error:", e)

# Execute if no exception occurs
else:
    print("Result =", result)

# Execute whether an exception occurs or not
finally:
    print("Thank you for using the calculator.")
```
