## Task 10: Variable-Length Arguments
Write a function total_marks(*marks) that accepts any number of marks and returns their sum.
Example:
Input:
45, 56, 89, 90

Output:
280
Bonus: Find the average as well.

```python
# Function to calculate total and average marks
def total_marks(*marks):
    total = 0

    # Calculate total marks
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)

    return total, average

# Get marks from the user
numbers = list(map(int, input("Enter marks separted by space: ").split()))

# Unpack the list and call the function
total, avg = total_marks(*numbers)

# Display the result
print(f"Total: {total}")
print(f"Average: {avg:.2f}")
```

## Task 12: Sort Tuples
Given:
students = [
   ("John", 80),
   ("Alice", 95),
   ("Bob", 72),
   ("David", 88)
]
Use a lambda function to sort the list based on marks.
Expected Output:
[
('Bob',72),
('John',80),
('David',88),
('Alice',95)
]

```python
# List of student names and marks
students = [
    ("John", 80),
    ("Alice", 95),
    ("Bob", 72),
    ("David", 88)
]

# Sort students by marks in ascending order
sorted_students = sorted(students, key=lambda x: x[1])

# Display the sorted list
print(sorted_students)
```

## Task 13: Filter Even Numbers
Given:
numbers = [10, 15, 22, 31, 40, 51]
Use lambda with filter() to print only even numbers.
Expected Output:
[10, 22, 40]

```python
# List of numbers
numbers = [10, 15, 22, 31, 40, 51]

# Filter only even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

# Display the result
print(even_numbers)
```

```python
# Function to display employee salary details
def display(employee_name, department, basic_salary, bonus):
    total_salary = basic_salary + bonus

    print(f"Employee: {employee_name}")
    print(f"Department: {department}")
    print(f"Basic Salary: {basic_salary}")
    print(f"Bonus: {bonus}")
    print(f"Total Salary: {total_salary}")

# Get input from the user
employee_name = input("Enter employee name: ")
department = input("Enter department: ")
basic_salary = float(input("Enter basic salary: "))
bonus = float(input("Enter bonus: "))

# Function call
display(employee_name, department, basic_salary, bonus)
```
