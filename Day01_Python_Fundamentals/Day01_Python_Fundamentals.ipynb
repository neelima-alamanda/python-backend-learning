# Task 6: Count Vowels
Take a string from the user and count how many vowels it contains.

```python
# Read a string from the user
value = input("Enter any string: ")

# Initialize the vowel counter
count = 0

# Check each character in the string
for item in value:

    # Increment the count if the character is a vowel
    if item in "aeiouAEIOU":
        count += 1

# Display the total number of vowels
print("Vowel Count:", count)
```

# Task 7: Reverse a String
Without using built-in reverse functions, print the string in reverse

```python
# Read a string from the user
string = input("Enter any string: ")

# Print the string in reverse order using a loop
for index in range(len(string) - 1, -1, -1):
    print(string[index], end="")
```

# Task 11: Check Palindrome
Check whether the entered string is a palindrome.

```python
# Read a string from the user
string = input("Enter a string: ")

# Reverse the string using slicing
rev = string[::-1]

# Display the reversed string
print("Reversed String:", rev)

# Check whether the original and reversed strings are the same
if string == rev:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

# Task 12: Username Validation
Take a username and check:
Starts with a letter
Contains no spaces
Length is at least 5
Print whether it is valid.

```python
# Read the username from the user
username = input("Enter username: ")

# Check if the username is empty
if username == "":
    print("Invalid")

# Validate the username
# Conditions:
# 1. First character should be an alphabet
# 2. Username should not contain spaces
# 3. Minimum length should be 5 characters
elif (username[0].isalpha()) and (" " not in username) and len(username) >= 5:
    print("Valid")

# Display invalid message if any condition fails
else:
    print("Not Valid")
```

# Task 13: Student Grade
Store marks in variables.
Calculate:
 Total
 Average
Using comparison operators, print:
 Distinction
 First Class
 Second Class
 Fail

```python
# Read marks of five subjects from the user
s1 = int(input("Enter Subject 1 marks: "))
s2 = int(input("Enter Subject 2 marks: "))
s3 = int(input("Enter Subject 3 marks: "))
s4 = int(input("Enter Subject 4 marks: "))
s5 = int(input("Enter Subject 5 marks: "))

# Calculate total marks
total = s1 + s2 + s3 + s4 + s5

# Calculate average marks
average = total / 5

# Display total and average
print("Total:", total)
print("Average:", average)

# Determine the grade based on the average
if average >= 90:
    print("Distinction")
elif average >= 80:
    print("First Class")
elif average >= 70:
    print("Second Class")
else:
    print("Fail")
```

# Task 17: Find the Second Largest Number

Write a Python program to find the **largest** and **second largest** numbers among three given numbers.


```python
# Given numbers
a = 10
b = 15
c = 20

# Find the largest and second largest numbers
if a > b and a > c:

    biggest = a

    if b > c:
        second_largest = b
    else:
        second_largest = c

elif b > a and b > c:

    biggest = b

    if a > c:
        second_largest = a
    else:
        second_largest = c

else:

    biggest = c

    if a > b:
        second_largest = a
    else:
        second_largest = b

# Display the result
print("Largest Number:", biggest)
print("Second Largest Number:", second_largest)
```
