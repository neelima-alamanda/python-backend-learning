## Task 1: Count Vowels
Write a function count_vowels(text) that returns the number of vowels in the given string.
Example:
Input:
"Programming"

Output:
3

```python
# Function to count the number of vowels in a word
def count_vowels(word):
    count = 0

    # Loop through each character in the word
    for ch in word:

        # Check whether the character is a vowel
        if ch in "aeiouAEIOU":
            count += 1

    # Return the total number of vowels
    return count


# Read input from the user
word = input("Enter a word to count vowels: ")

# Display the result
print("Number of vowels:", count_vowels(word))
```

## Task 2: Prime Number Checker
Write a function is_prime(number) that returns True if the number is prime; otherwise, return False.
Example:
Input:
17

Output:
True

```python
# Function to check whether a number is prime
def is_prime(num):

    # Numbers less than 2 are not prime
    if num < 2:
        return False

    # Check divisibility from 2 to num-1
    for i in range(2, num):

        # If divisible, it is not a prime number
        if num % i == 0:
            return False

    # If no divisors are found, it is prime
    return True


# Read input from the user
number = int(input("Enter a number: "))

# Display the result
if is_prime(number):
    print(number, "is a Prime Number")
else:
    print(number, "is Not a Prime Number")
```

## Task 3: Temperature Converter
Write a function convert_temperature(value, scale).
If scale = "C", convert Celsius to Fahrenheit.
If scale = "F", convert Fahrenheit to Celsius.
If any other value is given, return "Invalid Scale".
Example:
Input:
100, "C"

Output:
212.0
formulas:
f=c*9/5+32
c=f-32*5/9

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):

    # Apply conversion formula
    fahrenheit = (celsius * 9/5) + 32

    # Return converted temperature
    return fahrenheit


# Function to convert Fahrenheit to Celsius
def fahrenheit_to_celsius(fahrenheit):

    # Apply conversion formula
    celsius = (fahrenheit - 32) * 5/9

    # Return converted temperature
    return celsius


# Read Celsius temperature
celsius = float(input("Enter temperature in Celsius: "))
print(f"{celsius}°C = {celsius_to_fahrenheit(celsius):.2f}°F")

# Read Fahrenheit temperature
fahrenheit = float(input("Enter temperature in Fahrenheit: "))
print(f"{fahrenheit}°F = {fahrenheit_to_celsius(fahrenheit):.2f}°C")
```

## Task 4: Remove Duplicates
Write a function remove_duplicates(numbers) that returns a list containing only unique values while preserving the original order.
Example:
Input:
[2, 5, 2, 8, 5, 10, 8]

Output:
[2, 5, 8, 10]

```python
# Function to remove duplicate elements
def remove_duplicates(numbers):

    # Store unique elements
    unique_numbers = []

    # Traverse through each element
    for num in numbers:

        # Add element only if it is not already present
        if num not in unique_numbers:
            unique_numbers.append(num)

    # Return list with unique elements
    return unique_numbers


# Read list from the user
numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

# Display the result
print("List after removing duplicates:", remove_duplicates(numbers))
```

## Task 5: Second Largest Number
Write a function second_largest(numbers) that returns the second largest unique number in a list.
Example:
Input:
[10, 5, 20, 8, 20, 15]

Output:
15

```python
# Function to find the second largest number
def second_largest(numbers):

    # Remove duplicate values and sort the list
    new_list = sorted(set(numbers))

    # Check if at least two unique numbers are present
    if len(new_list) >= 2:
        return new_list[-2]

    # Return message if second largest element does not exist
    else:
        return "Second largest element does not exist."


# Input list
numbers = [10, 990]

# Display the result
print("Second Largest Number:", second_largest(numbers))
```

## Task 6: Password Strength Checker
Write a function check_password(password).
Return:
"Strong" if the password
has at least 8 characters,
contains at least one uppercase letter,
one lowercase letter,
and one digit.
Otherwise return "Weak".

```python
# Function to check the strength of a password
def check_password(password):

    # Initialize flags
    upper = lower = digit = False

    # Check if password length is at least 8 characters
    if len(password) >= 8:

        # Traverse each character in the password
        for ch in password:

            # Check for uppercase letter
            if ch.isupper():
                upper = True

            # Check for lowercase letter
            if ch.islower():
                lower = True

            # Check for digit
            if ch.isdigit():
                digit = True

        # Password is strong if all conditions are satisfied
        if upper and lower and digit:
            return "Strong"

    # Otherwise, password is weak
    return "Weak"


# Read password from the user
password = input("Enter password: ")

# Display the result
print("Password Strength:", check_password(password))
```

## Task 7: Student Result Generator
Create a function:
student_result(name, marks, passing_marks=35)
The function should:
Print the student's name.
Print whether the student passed or failed.
Display the result using an f-string.
Call the function using:
positional arguments
keyword arguments
default argument

```python
# Function to check student result
def student_result(name, marks, passing_marks=35):

    # Assume the student has failed
    status = "Failed"

    # Check whether the student has passed
    if marks >= passing_marks:
        status = "Passed"

    # Display the student's result
    print(f"Student Name : {name}")
    print(f"Status       : {status}")


# Function call using positional arguments
print("Using Positional Arguments:")
student_result("Sun", 40)

print()

# Function call using keyword arguments
print("Using Keyword Arguments:")
student_result(name="Moon", marks=7)
```

## Task 8: Shopping Bill (*args)
Write a function:
shopping_bill(customer_name, *prices)
The function should:
Calculate the total bill.
Find the most expensive item.
Find the cheapest item.
Display all details using f-strings.

```python
# Function to calculate the shopping bill
def shopping_bill(customer_name, *prices):

    # Check whether any prices are provided
    if not prices:
        print("No items purchased.")
        return

    # Find the most expensive item
    expensive = max(prices)

    # Find the cheapest item
    cheap = min(prices)

    # Calculate the total bill
    total_bill = sum(prices)

    # Display the shopping bill
    print(f"Customer Name : {customer_name}")
    print(f"Total Bill    : {total_bill}")
    print(f"Expensive Item: {expensive}")
    print(f"Cheapest Item : {cheap}")


# Function call
shopping_bill("Anu", 30, 3000, 684, 990)
```

## Task 9: Employee Details (**kwargs)
Write a function:
employee_details(**details)
Print all the key-value pairs neatly.
Example:
employee_details(
   name="John",
   department="HR",
   salary=45000,
   city="Hyderabad"
)

```python
# Function to display employee details
def employee_details(**details):

    # Traverse through all key-value pairs
    for key, value in details.items():

        # Display each employee detail
        print(f"{key} : {value}")


# Function call with keyword arguments
employee_details(
    name="John",
    department="HR",
    salary=45000,
    city="Hyderabad"
)
```

## Task 10: Mixed Arguments
Write a function:
calculate_marks(name, *marks, bonus=0)
The function should:
Calculate total marks.
Add bonus marks.
Calculate average.
Print all details using f-strings.

```python
# Function to calculate total marks and average
def calculate_marks(name, *marks, bonus=1):

    # Check whether marks are provided
    if not marks:
        print("No marks entered.")
        return

    # Calculate total marks
    total_marks = sum(marks)

    # Add bonus marks
    total_marks += bonus

    # Calculate average marks
    avg = total_marks / len(marks)

    # Display student details
    print(f"Student Name : {name}")
    print(f"Total Marks  : {total_marks}")
    print(f"Average      : {avg:.2f}")


# Function call
calculate_marks("Sri", 4, 5, 7, 6, bonus=10)
```

## Task 11: Sort Dictionary
Given:
employees = {
   "John": 45000,
   "Alice": 52000,
   "Bob": 40000,
   "David": 60000
}
Convert the dictionary into a list of tuples and sort it by salary using a lambda function.

```python
# Dictionary containing employee names and their salaries
employees = {
    "John": 45000,
    "Alice": 52000,
    "Bob": 40000,
    "David": 60000
}

# Sort employees based on salary in ascending order
sorted_employees = sorted(employees.items(), key=lambda x: x[1])

# Display the sorted list of employees
print("Employees sorted by salary:")
print(sorted_employees)
```

## Task 12: Filter Long Words
Given:
words = ["python", "AI", "machine", "data", "science", "ML"]
Use filter() with a lambda function to print only the words having more than 5 characters.
Expected Output
['python', 'machine', 'science']

```python
# List of words
words = ["python", "AI", "machine", "data", "science", "ML"]

# Filter words having more than 5 characters
long_words = list(filter(lambda word: len(word) > 5, words))

# Display the filtered words
print("Words with more than 5 characters:")
print(long_words)
```

## Task 13: Price After Discount
Given:
prices = [500, 1200, 800, 1500]
Use map() with a lambda function to apply a 10% discount to every price.
Expected Output:
[450.0, 1080.0, 720.0, 1350.0]

```python
# List of product prices
prices = [500, 1200, 800, 1500]

# Apply a 10% discount to each price using map() and lambda
discounted_prices = list(map(lambda price: price - (0.10 * price), prices))

# Display the discounted prices
print("Prices after applying 10% discount:")
print(discounted_prices)
```

## Task 14: Student Report Card
Create variables:
Name
Roll Number
Marks in 5 subjects
Calculate:
Total
Average
Grade
Display the report card in a well-formatted layout using only f-strings.
Example:
----------------------------------
Student Report Card
----------------------------------
Name        : Greeshma
Roll No     : 24

Total Marks : 445
Average     : 89.0
Grade       : A

```python
# Read student details
name = input("Enter student name: ")
roll_number = int(input("Enter student Roll No.: "))

# Read marks of 5 subjects
marks = list(map(int, input("Enter 5 subject marks (comma separated): ").split(",")))

# Calculate total marks
total = sum(marks)

# Calculate average marks
average = total / len(marks)

# Assign grade based on total marks
if total >= 400:
    grade = "A"
elif total >= 300:
    grade = "B"
else:
    grade = "C"

# Display the student report card
print("-------------------------------------")
print("        Student Report Card")
print("-------------------------------------")
print(f"Name         : {name}")
print(f"Roll No.     : {roll_number}")
print(f"Total Marks  : {total}")
print(f"Average      : {average:.2f}")
print(f"Grade        : {grade}")
print("-------------------------------------")
```

## Task 15: Global Variable Counter
Create a global variable:
count = 0
Write a function increment() that increases the value of count by 1 every time it is called.
Call the function five times.
Questions:
What happens if you don't use the global keyword?
Why is the global keyword required here?
What is the final value of count after five function calls?

```python
# Global variable
count = 0

# Function to increment the global counter
def increment():

    # Access the global variable
    global count

    # Increase the counter by 1
    count += 1

    # Display the updated count
    print(f"Counter Value: {count}")


# Function calls
increment()
increment()
increment()
increment()
increment()
```
